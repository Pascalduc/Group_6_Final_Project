```python
# Import our dependencies
#from sklearn.model_selection import train_test_split
#from sklearn.preprocessing import StandardScaler,OneHotEncoder
import pandas as pd
```

```python
# Import and read the Air BnB Listings dataset.
listings_df = pd.read_csv("Resources/Listings_mockup.csv")
listings_df.head(10)
```

```python
# Import and read the Air BnB Reviews dataset.
reviews_df = pd.read_csv("Resources/Reviews_mockup.csv")
reviews_df.head(10)
```

```python
# Merging the comments from the reviews_df with the listings_df.
airbnb_df = pd.merge(listings_df, reviews_df, how='left', left_on=['id'], right_on=['listing_id'])
airbnb_df.head(10)
```

```python
airbnb_df.columns
```

```python
airbnb_df.drop(['listing_id', 'date'], axis=1, inplace=True)
airbnb_df
```
